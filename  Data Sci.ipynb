```python
plt.figure(figsize=(8, 6))
sns.histplot(data['Salary'], bins=20, kde=True)
plt.title('Salary Distribution')
plt.show()
```

```python
avg_salary_by_title = data.groupby('Job Title')['Salary'].mean().sort_values(ascending=False)
print(avg_salary_by_title)
```

```python
avg_salary_by_company = data.groupby('Company')['Salary'].mean().sort_values(ascending=False)
print(avg_salary_by_company)
```

```python
jobs_by_city = data['City'].value_counts()
print(jobs_by_city)
```

```python
avg_salary_by_city = data.groupby('City')['Salary'].mean().sort_values(ascending=False)
print(avg_salary_by_city)
```

```python
avg_salary_by_country = data.groupby('Country')['Salary'].mean().sort_values(ascending=False)
print(avg_salary_by_country)
```

```python
corr = data[['Company Score', 'Salary']].corr()
print(corr)
```

```python
plt.figure(figsize=(6, 4))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation between Company Score and Salary')
plt.show()
```

```python
avg_salary_by_remote = data.groupby('Remote')['Salary'].mean()
print(avg_salary_by_remote)
```

Conclusion : No big difference are remote and outsite job

```python
jobs_by_remote = data['Remote'].value_counts()
print(jobs_by_remote)
```

```python
sns.countplot(x=data["Remote"])
plt.show()
```

Conclusion : Maximum company are the onsite job .
             very less number of company are remote side job

```python
data['Skills'] = data['Skills'].str.split(', ')
skills_exploded = data.explode('Skills')
top_skills = skills_exploded['Skills'].value_counts().head(10)
print(top_skills)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=top_skills.values, y=top_skills.index, palette='viridis')
plt.title('Top 10 Most Required Skills')
plt.show()
```

Conclusion :Most requied skill is Machine Learning
